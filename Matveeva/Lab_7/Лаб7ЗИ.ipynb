```python
import random

def generate_keys(p, g):
    # Генерация приватного ключа
    private_key = random.randint(1, p-2)
    # Генерация публичного ключа
    public_key = pow(g, private_key, p)
    return private_key, public_key

def encrypt(message, public_key, p, g):
    # Генерация случайного числа
    k = random.randint(1, p-2)
    # Вычисление шифротекста
    c1 = pow(g, k, p)
    c2 = [(ord(char) - ord('0')) * pow(public_key, k, p) % p for char in message]
    return c1, c2

def decrypt(c1, c2, private_key, p):
    # Вычисление общего секрета
    s = pow(c1, private_key, p)
    # Восстановление исходного сообщения
    s_inv = pow(s, p-2, p)  # Обратное значение по модулю
    decrypted_message = ''.join(chr((char * s_inv) % p + ord('0')) for char in c2)
    return decrypted_message

# Основная программа
if __name__ == "__main__":
    # Исходные данные
    message = "344865"
    p = 23  # Простое число
    g = 7   # Генератор

    # Генерация ключей
    private_key, public_key = generate_keys(p, g)

    # Шифровка сообщения вариант 9
    c1, c2 = encrypt(message, public_key, p, g)

    # Запись шифротекста в файл
    with open("ЗашифрованныйТекст.txt", "w") as f:
        f.write(f"{c1}\n")
        f.write(" ".join(map(str, c2)))

    # Расшифровка сообщения
    decrypted_message = decrypt(c1, c2, private_key, p)

    # Запись восстановленного сообщения в файл
    with open("ВостановленноеСообщение.txt", "w") as f:
        f.write(decrypted_message)

    print(f"Исходное сообщение: {message}")
    print(f"Шифротекст: {c1}, {c2}")
    print(f"Восстановленное сообщение: {decrypted_message}")
```
