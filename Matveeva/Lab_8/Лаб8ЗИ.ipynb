```python
# Импортируем библиотеку
import os

# Функция для возведения в степень с модулем
def power_mod(base, exp, mod):
    result = 1
    base = base % mod
    while exp > 0:
        if (exp % 2) == 1:  # Если exp нечетное
            result = (result * base) % mod
        exp = exp >> 1  # Делим exp на 2
        base = (base * base) % mod
    return result

# Функция для шифровки сообщения
def encrypt(message, e, n):
    encrypted_message = []
    for char in message:
        # Преобразуем символ в число и шифруем
        encrypted_char = power_mod(int(char), e, n)
        encrypted_message.append(encrypted_char)
    return encrypted_message

# Функция для расшифровки сообщения
def decrypt(encrypted_message, d, n):
    decrypted_message = ''
    for num in encrypted_message:
        # Расшифровываем число и преобразуем обратно в символ
        decrypted_char = str(power_mod(num, d, n))
        decrypted_message += decrypted_char
    return decrypted_message

# Основные параметры RSA
P = 3
Q = 11
N = P * Q
E = 7      # Открытый ключ
D = 3      # Закрытый ключ

# Исходное сообщение
original_message = "1234567890"

# Шифруем сообщение
encrypted_message = encrypt(original_message, E, N)

# Сохраняем зашифрованное сообщение в файл
with open('encrypted_message.txt', 'w') as f:
    f.write(' '.join(map(str, encrypted_message)))

# Расшифровываем сообщение
decrypted_message = decrypt(encrypted_message, D, N)

# Сохраняем восстановленное сообщение в файл
with open('decrypted_message.txt', 'w') as f:
    f.write(decrypted_message)

print("Исходное сообщение:", original_message)
print("Зашифрованное сообщение:", encrypted_message)
print("Восстановленное сообщение:", decrypted_message)
```
