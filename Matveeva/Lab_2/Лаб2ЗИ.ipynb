```python
def vigenere_encrypt(plaintext, key):
    encrypted_text = []
    key_length = len(key)
    key_as_int = [ord(i) - ord('А') for i in key.upper() if i in 'АБВГДЕЁЖЗИЙКЛМНОПРСТУФХЦЧШЩЪЫЬЭЮЯ']
    plaintext_int = [ord(i) - ord('А') for i in plaintext.upper() if i in 'АБВГДЕЁЖЗИЙКЛМНОПРСТУФХЦЧШЩЪЫЬЭЮЯ']

    for i in range(len(plaintext_int)):
        value = (plaintext_int[i] + key_as_int[i % key_length]) % 33  # 33 - количество букв в русском алфавите
        encrypted_text.append(chr(value + ord('А')))

    return ''.join(encrypted_text)

def vigenere_decrypt(ciphertext, key):
    decrypted_text = []
    key_length = len(key)
    key_as_int = [ord(i) - ord('А') for i in key.upper() if i in 'АБВГДЕЁЖЗИЙКЛМНОПРСТУФХЦЧШЩЪЫЬЭЮЯ']
    ciphertext_int = [ord(i) - ord('А') for i in ciphertext.upper() if i in 'АБВГДЕЁЖЗИЙКЛМНОПРСТУФХЦЧШЩЪЫЬЭЮЯ']

    for i in range(len(ciphertext_int)):
        value = (ciphertext_int[i] - key_as_int[i % key_length]) % 33  # 33 - количество букв в русском алфавите
        decrypted_text.append(chr(value + ord('А')))

    return ''.join(decrypted_text)

# Ввод данных с клавиатуры
plaintext = input('Введите открытый текст: ')
key = input('Введите ключ: ')

encrypted = vigenere_encrypt(plaintext, key)
print(f'Зашифрованное сообщение: {encrypted}')

decrypted = vigenere_decrypt(encrypted, key)
print(f'Расшифрованное сообщение: {decrypted}')
```
