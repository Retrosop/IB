```python
import random

class InfiniteKeyGenerator:
    # Генератор бесконечной последовательности ключей
    def __init__(self):
        self.previous = random.randint(0, 255)  # Начальное значение для генерации

    def next_key(self):
        # Возвращает следующий ключ по методу аддитивного генератора
        self.previous = (self.previous + random.randint(0, 255)) % 256
        return self.previous

def vernam_encrypt(message, key_generator):
    # Шифрует сообщение с помощью бесконечного ключа методом Вернама
    return bytes([m ^ key_generator.next_key() for m in message])

def vernam_decrypt(ciphertext, key_generator):
    # Дешифрует сообщение с помощью бесконечного ключа методом Вернама
    return bytes([c ^ key_generator.next_key() for c in ciphertext])

# Пример использования
if __name__ == "__main__":
    message = input("Введите сообщение для шифрования: ")  # Ввод сообщения с клавиатуры
    message_bytes = message.encode('utf-8')  # Преобразуем строку в байты
    key_generator = InfiniteKeyGenerator()  # Создаем генератор ключей

    # Шифрование
    ciphertext = vernam_encrypt(message_bytes, key_generator)
    print("Зашифрованное сообщение:", ciphertext)

    # Дешифрование с тем же генератором ключей
    decrypted_message = vernam_decrypt(ciphertext, key_generator)

    # Выводим зашифрованное и расшифрованное сообщение в байтах
    print("Расшифрованное сообщение (в байтах):", decrypted_message)
```
