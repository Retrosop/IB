```python
import math
import random
import sys

# Настройка кодировки ввода на UTF-8
sys.stdin.reconfigure(encoding='utf-8')

def gcd(m, n):
    # Функция для нахождения наибольшего общего делителя (НОД) двух чисел
    while m != 0 and n != 0:
        if m > n:
            m = m % n
        else:
            n = n % m
    return n if n != 0 else m

def is_prime(p):
    # Функция для проверки, является ли число простым
    if p <= 1:
        return False
    for k in range(2, int(math.sqrt(p)) + 1):
        if p % k == 0:
            return False
    return True

def next_prime(p):
    # Функция для нахождения следующего простого числа, начиная с p
    while not is_prime(p):
        p += 1
    return p

def power_mod(g, k, p):
    # Функция для вычисления (g^k) mod p
    return pow(g, k, p)

def elgamal():
    p = 11  # Начальное простое число
    g = 3   # Генератор
    a = 6   # Секретный ключ
    p = next_prime(p)  # Находим следующее простое число
    y = power_mod(g, a, p)  # Вычисляем открытый ключ y
    print(f'Открытый ключ - y = {y}')

    # Ввод пароля с подтверждением
    while True:
        password = input("Введите пароль для шифрования: ").encode('utf-8')  # Ввод пароля
        confirm_password = input("Подтвердите пароль: ").encode('utf-8')  # Подтверждение пароля
        if password.decode('utf-8') == confirm_password.decode('utf-8'):
            break  # Если пароли совпадают, выходим из цикла
        else:
            print("Пароли не совпадают. Пожалуйста, попробуйте снова.")

    # Шифрование пароля
    encrypted_password = []
    for char in password.decode('utf-8'):  # Декодируем пароль для обработки
        k = random.randint(1, p - 2)  # Генерируем случайное число k
        y1 = chr(power_mod(g, k, p))  # Вычисляем y1
        key = power_mod(y, k, p)  # Вычисляем ключ
        y2 = chr(ord(char) ^ key)  # Шифруем символ пароля
        encrypted_password.append((y1, y2))  # Добавляем зашифрованный символ в список

    # Добавляем зашифрованный пароль в начало зашифрованного сообщения
    encrypted_message = []
    message = input("Введите сообщение для шифрования: ").encode('utf-8')  # Ввод сообщения
    for c in message.decode('utf-8'):  # Декодируем сообщение для обработки
        k1 = random.randint(1, p - 2)  # Генерируем случайное число k1
        k = k1
        while gcd(k, p - 1) != 1 and k < p - 1 or k == 31:  # Находим подходящее k
            k += 1
        if k == 1:
            k = k1
            while gcd(k, p - 1) != 1:
                k -= 1
            if k == p - 1:
                print('Ошибка!!! Нельзя найти подходящее k')
                return

        y1 = chr(power_mod(g, k, p))  # Вычисляем y1 для символа сообщения
        key = power_mod(y, k, p)  # Вычисляем ключ
        y2 = chr(ord(c) ^ key)  # Шифруем символ сообщения
        encrypted_message.append((y1, y2))  # Добавляем зашифрованный символ в список

    # Выводим зашифрованное сообщение
    print("Зашифрованный пароль:", encrypted_password)
    print("Зашифрованное сообщение:", encrypted_message)

    # Дешифрование пароля
    decrypted_password = []
    for y1, y2 in encrypted_password:
        c = chr((power_mod(ord(y1), a, p) ^ ord(y2)))  # Дешифруем символ пароля
        decrypted_password.append(c)  # Добавляем расшифрованный символ в список

    # Дешифрование сообщения
    decrypted_message = []
    for y1, y2 in encrypted_message:
        c = chr((power_mod(ord(y1), a, p) ^ ord(y2)))  # Дешифруем символ сообщения
        decrypted_message.append(c)  # Добавляем расшифрованный символ в список

    # Выводим расшифрованные данные
    print("Расшифрованный пароль:", ''.join(decrypted_password))  # Вывод расшифрованного пароля
    print("Расшифрованное сообщение:", ''.join(decrypted_message))  # Вывод расшифрованного сообщения

if __name__ == '__main__':
    elgamal()  # Запуск функции ElGamal
```
