```python
from Crypto.Cipher import DES
from Crypto.Util.Padding import pad, unpad
import os

def generate_key():
    """Генерация случайного ключа для DES."""
    return os.urandom(8)  # DES использует 64-битный ключ (8 байт)

def encrypt(plain_text, key):
    """Шифрование текста с использованием DES в режиме ECB."""
    cipher = DES.new(key, DES.MODE_ECB)
    padded_text = pad(plain_text.encode(), DES.block_size)
    encrypted_text = cipher.encrypt(padded_text)
    return encrypted_text

def decrypt(encrypted_text, key):
    """Расшифрование текста с использованием DES в режиме ECB."""
    cipher = DES.new(key, DES.MODE_ECB)
    decrypted_padded_text = cipher.decrypt(encrypted_text)
    return unpad(decrypted_padded_text, DES.block_size).decode()

# Пример использования
if __name__ == "__main__":
    key = generate_key()
    print(f"Сгенерированный ключ: {key.hex()}")

    plain_text = input("Введите текст для шифрования: ")
    print(f"Исходный текст: {plain_text}")

    encrypted_text = encrypt(plain_text, key)
    print(f"Зашифрованный текст: {encrypted_text.hex()}")

    decrypted_text = decrypt(encrypted_text, key)
    print(f"Расшифрованный текст: {decrypted_text}")
```

```python
from Crypto.Cipher import DES
from Crypto.Util.Padding import pad, unpad

def triple_des_encrypt(plaintext, key1, key2, key3):
    # Убедимся, что открытый текст кратен 8 байтам (размер блока DES)
    padded_text = pad(plaintext, DES.block_size)

    # Первое шифрование с помощью ключа key1
    cipher1 = DES.new(key1, DES.MODE_ECB)
    encrypted1 = cipher1.encrypt(padded_text)

    # Дешифрование с помощью ключа key2
    cipher2 = DES.new(key2, DES.MODE_ECB)
    decrypted2 = cipher2.decrypt(encrypted1)

    # Второе шифрование с помощью ключа key3
    cipher3 = DES.new(key3, DES.MODE_ECB)
    encrypted3 = cipher3.encrypt(decrypted2)

    return encrypted3

def triple_des_decrypt(ciphertext, key1, key2, key3):
    # Первое дешифрование с помощью ключа key3
    cipher3 = DES.new(key3, DES.MODE_ECB)
    decrypted3 = cipher3.decrypt(ciphertext)

    # Шифрование с помощью ключа key2
    cipher2 = DES.new(key2, DES.MODE_ECB)
    encrypted2 = cipher2.encrypt(decrypted3)

    # Финальное дешифрование с помощью ключа key1
    cipher1 = DES.new(key1, DES.MODE_ECB)
    decrypted1 = cipher1.decrypt(encrypted2)

    # Убираем паддинг из расшифрованного текста
    return unpad(decrypted1, DES.block_size)

# Ввод данных с клавиатуры
key1 = b'12345678'  # 8-байтовый ключ для DES
key2 = b'abcdefgh'  # 8-байтовый ключ для DES
key3 = b'87654321'  # 8-байтовый ключ для DES
plaintext = input("Введите открытый текст: ").encode('utf-8')

if len(key1) != 8 or len(key2) != 8 or len(key3) != 8:
    print("Ключи должны состоять ровно из 8 символов!")
    # Шифруем открытый текст
else:
    ciphertext = triple_des_encrypt(plaintext, key1, key2, key3)
    print("Зашифрованный текст:", ciphertext)

    # Дешифруем зашифрованный текст
    decrypted_text = triple_des_decrypt(ciphertext, key1, key2, key3)
    print("Расшифрованный текст:", decrypted_text.decode('utf-8'))
```
