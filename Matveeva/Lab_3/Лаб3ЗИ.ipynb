```python
class MultiplicativeGenerator:
    def __init__(self, a, m, y0):
        self.a = a
        self.m = m
        self.y = y0

    def next(self):
        self.y = (self.a * self.y) % self.m
        return self.y

def generate_gamma(length, a, m, y0):
    generator = MultiplicativeGenerator(a, m, y0)
    return [generator.next() for _ in range(length)]

def encrypt_decrypt(text, gamma):
    result = []
    for i in range(len(text)):
        # Шифрование/дешифрование с использованием XOR
        result.append(chr(ord(text[i]) ^ (gamma[i] % 256)))
    return ''.join(result)

# Параметры генератора
a = 7
m = 4096
y0 = 502

# Ввод текста с клавиатуры
text = input("Введите текст для шифрования: ")
gamma = generate_gamma(len(text), a, m, y0)

# Шифрование
encrypted_text = encrypt_decrypt(text, gamma)
print("Зашифрованный текст:", encrypted_text)

# Дешифрование
decrypted_text = encrypt_decrypt(encrypted_text, gamma)
print("Расшифрованный текст:", decrypted_text)
```
